```python
# Turning into a dataframe because easiest way I can find online to pretty format this 
import pandas as pd
from pickle import load
from nltk import download
from nltk.corpus import stopwords
```

```python
with open('exploratory_table.pickle', 'rb') as handle:
    table = load(handle)

df = df = pd.DataFrame.from_dict(table, orient='index')
```

```python
df = df.sort_values(by=['total'], ascending=False)

# Filter out stopwords using nltk
stop = set(stopwords.words("english"))
no_stopwords_df = df.loc[~df.index.isin(stop)]

no_stopwords_df.iloc[0:30]
```
