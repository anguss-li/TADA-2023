```python
# Turning into a dataframe because easiest way I can find online to pretty format this 
import pandas as pd
import numpy as np
from pickle import load
from nltk import download
from nltk.corpus import stopwords
```

```python
with open("exploratory_table.pickle", "rb") as handle:
    table = load(handle)

df = pd.DataFrame.from_dict(table, orient="index")
```

```python
with open("confidence_intervals.pickle", "rb") as handle:
    CI = load(handle)
```

```python
CI = pd.Series(CI)
df["CI (90%)"] = CI
```

```python
split_CI = pd.DataFrame(df["CI (90%)"].to_list(), columns=["min","max"], index=df.index)
df["Significant"] = np.where((0 < split_CI["min"]), True, False)
```

```python
df = df.sort_values(by=["PPMI"], ascending=False)

# Filter out stopwords using nltk and words not used by female speakers
stop = set(stopwords.words("english"))
filtered_df = df.loc[(~df.index.isin(stop)) & (df["PPMI"] != None)]

# We can set a cutoff for a good PPMI to explore 
# filtered_df = filtered_df.loc[df["PPMI"] < 0.7]
filtered_df = filtered_df.loc[(df["M count"] > 20) & (df["F count"] > 20)]
# filtered_df = filtered_df.loc[df["Significant"] == True]

filtered_df[0:30]
```
